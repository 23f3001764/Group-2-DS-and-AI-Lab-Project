# Importing Data and checking structure

```python
import gdown

file_id = "1dnfyP7dHdYYRFPmu4JxZaWgaCt43KsMw"
url = f"https://drive.google.com/uc?id={file_id}"

gdown.download(url, "data.zip", quiet=False, fuzzy=True)
```

```python
import zipfile

with zipfile.ZipFile("data.zip", 'r') as zip_ref:
    zip_ref.extractall("/kaggle/working/data")
```

```python
file_path = "/kaggle/working/data/nutrivision_merged/data.yaml"

with open(file_path, "r") as f:
    data = f.read()

data = data.replace(
    "/content/nutrivision_merged",
    "/kaggle/working/data/nutrivision_merged"
)

with open(file_path, "w") as f:
    f.write(data)

print("YAML fixed ✅")
```

# Training and Testing

```python
!pip install ultralytics
from ultralytics import YOLO
model = YOLO("yolo26n.pt")

model.train(
    data="/kaggle/working/data/nutrivision_merged/data.yaml",
    epochs=30,
    patience=15,
    imgsz=640,
    batch=8)
```

```python
!ls /kaggle/working/runs/detect/train/weights
```

```python
from ultralytics import YOLO

models = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

metrics = models.val()
```

```python
print("mAP@50:", metrics.box.map50)
print("mAP@50-95:", metrics.box.map)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```

```python
from ultralytics import YOLO

model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

metrics_test = model.val(
    data="/kaggle/working/data/nutrivision_merged/data.yaml",
    split="test"
)
```

```python
print("mAP@50:", metrics_test.box.map50)
print("mAP@50-95:", metrics_test.box.map)
print("Precision:", metrics_test.box.mp)
print("Recall:", metrics_test.box.mr)
```
