```python
def load_and_normalize(yaml_path):
    with open(yaml_path) as f:
        data = yaml.safe_load(f)
    names = data['names']
    # Handle Roboflow dict format: {0: 'name', 1: 'name'}
    if isinstance(names, dict):
        names = [names[i] for i in sorted(names.keys())]
    normalized = [normalize(n) for n in names]
    return normalized

norm_ds1 = load_and_normalize("/content/FOOD-DETECTION-3/data.yaml")
norm_ds2 = load_and_normalize("/content/Indian-Food-Yolov5-3/data.yaml")
norm_ds3 = load_and_normalize("/content/IndianFoodNet-1/data.yaml")

print("DS1 normalized:", norm_ds1)
print(f"\nDS2 normalized:", norm_ds2)
print(f"\nDS3 normalized:", norm_ds3)
```

```python
all_classes = sorted(set(norm_ds1 + norm_ds2 + norm_ds3))

print(f"✅ Total unified classes: {len(all_classes)}\n")
for i, c in enumerate(all_classes):
    print(f"  {i:3d}: {c}")
```

```python
def build_remap(normalized_list, master):
    return {
        old_id: master.index(name)
        for old_id, name in enumerate(normalized_list)
    }

remap_ds1 = build_remap(norm_ds1, all_classes)
remap_ds2 = build_remap(norm_ds2, all_classes)
remap_ds3 = build_remap(norm_ds3, all_classes)

print("DS1 remap:", remap_ds1)
print("\nDS2 remap:", remap_ds2)
print("\nDS3 remap:", remap_ds3)
```

```python
def audit_labels(ds_path, remap_dict, ds_name):
    ds_path = Path(ds_path)
    bad = []
    for split in ['train', 'valid', 'test']:
        lbl_dir = ds_path / split / 'labels'
        if not lbl_dir.exists():
            continue
        for lbl_file in lbl_dir.glob("*.txt"):
            with open(lbl_file) as f:
                for line in f:
                    parts = line.strip().split()
                    if not parts:
                        continue
                    cid = int(parts[0])
                    if cid not in remap_dict:
                        bad.append((split, lbl_file.name, cid))

    if bad:
        print(f"⚠️  {ds_name}: {len(bad)} corrupt annotations found (will be skipped):")
        for split, fname, cid in bad[:10]:
            print(f"     [{split}] {fname} → unknown class ID {cid}")
    else:
        print(f"✅  {ds_name}: All label IDs are valid")

audit_labels("/content/FOOD-DETECTION-3",     remap_ds1, "DS1")
audit_labels("/content/Indian-Food-Yolov5-3", remap_ds2, "DS2")
audit_labels("/content/IndianFoodNet-1",       remap_ds3, "DS3")
```

```python
MERGED = Path("/content/nutrivision_merged")

# Clean slate
if MERGED.exists():
    shutil.rmtree(MERGED)

for split in ['train', 'valid', 'test']:
    (MERGED / split / 'images').mkdir(parents=True, exist_ok=True)
    (MERGED / split / 'labels').mkdir(parents=True, exist_ok=True)

print("📁 Output folder created\n")


def remap_label_file(src, dst, remap):
    with open(src) as f:
        lines = f.readlines()
    out = []
    skipped = 0
    for line in lines:
        parts = line.strip().split()
        if not parts:
            continue
        old_id = int(parts[0])
        if old_id not in remap:
            skipped += 1
            continue                      # skip corrupt annotations
        out.append(f"{remap[old_id]} " + " ".join(parts[1:]))
    with open(dst, 'w') as f:
        f.write("\n".join(out))
    return skipped


def merge_dataset(src_dir, remap, prefix):
    src_dir = Path(src_dir)
    total_imgs, total_skipped = 0, 0

    for split in ['train', 'valid', 'test']:
        img_src = src_dir / split / 'images'
        lbl_src = src_dir / split / 'labels'

        if not img_src.exists():
            print(f"  ⚠️  {prefix}/{split} — not found, skipping")
            continue

        img_dst = MERGED / split / 'images'
        lbl_dst = MERGED / split / 'labels'
        count = 0

        for img in img_src.iterdir():
            if img.suffix.lower() not in ['.jpg', '.jpeg', '.png']:
                continue

            # Unique filename to avoid collisions across datasets
            new_stem = f"{prefix}_{img.stem}"
            shutil.copy2(img, img_dst / (new_stem + img.suffix))

            lbl = lbl_src / (img.stem + ".txt")
            if lbl.exists():
                skipped = remap_label_file(
                    lbl,
                    lbl_dst / (new_stem + ".txt"),
                    remap
                )
                total_skipped += skipped
            else:
                # Keep image but write empty label (background/unannotated)
                open(lbl_dst / (new_stem + ".txt"), 'w').close()

            count += 1

        total_imgs += count
        print(f"  ✅ {prefix}/{split:5s} → {count} images")

    if total_skipped:
        print(f"  ⚠️  {prefix}: {total_skipped} corrupt annotations skipped")
    print()


merge_dataset("/content/FOOD-DETECTION-3",     remap_ds1, "ds1")
merge_dataset("/content/Indian-Food-Yolov5-3", remap_ds2, "ds2")
merge_dataset("/content/IndianFoodNet-1",       remap_ds3, "ds3")
```
