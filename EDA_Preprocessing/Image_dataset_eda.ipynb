# Exploratory Data Analysis (EDA) – Image Datasets

This notebook performs exploratory data analysis on the image datasets used in the NutriVision project. The goal is to understand dataset characteristics, identify potential data quality issues, and prepare the datasets for model training.

Datasets analyzed:
- Khana Dataset (Indian cuisine classification dataset)
- Food20 Dataset (20-class Indian food dataset)
- IndianFood10 Dataset (object detection dataset with bounding box annotations)

## Import Required Libraries

```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm import tqdm
```

## Define Dataset Paths

```python
khana_path = "/content/khana_dataset/khana"
food20_path = "/content/food20dataset/train_set"
roboflow_path = "/content/indianfood10-1"
```

## Khana Dataset Overview

The Khana dataset contains a large collection of Indian cuisine images organized into 80 food classes. This dataset is used primarily for large-scale food classification and feature learning.

```python
khana_classes = os.listdir(khana_path)

print("Total classes:", len(khana_classes))
print("Sample classes:", khana_classes[:10])
```

### Image Distribution Across Classes (Khana Dataset)

```python
class_counts = {}

for cls in khana_classes:
    class_folder = os.path.join(khana_path, cls)
    if os.path.isdir(class_folder):
        class_counts[cls] = len(os.listdir(class_folder))

df_khana = pd.DataFrame(list(class_counts.items()), columns=["Food Class","Image Count"])
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Food Class", y="Image Count",
            data=df_khana.sort_values("Image Count", ascending=False).head(20))

plt.xticks(rotation=90)
plt.title("Top 20 Classes by Image Count – Khana Dataset")
plt.show()
```

## Food20 Dataset Overview

The Food20 dataset contains approximately 2,000 images across 20 Indian food classes. The dataset is used to increase diversity and support model generalization.

```python
food20_classes = os.listdir(food20_path)

print("Number of classes:", len(food20_classes))
print("Classes:", food20_classes)
```

### Image Distribution Across Classes (Food20 Dataset)

```python
food20_counts = {}

for cls in food20_classes:
    class_folder = os.path.join(food20_path, cls)
    food20_counts[cls] = len(os.listdir(class_folder))

df_food20 = pd.DataFrame(list(food20_counts.items()),
                         columns=["Food Class","Image Count"])
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Food Class", y="Image Count", data=df_food20)

plt.xticks(rotation=90)
plt.title("Class Distribution – Food20 Dataset")
plt.show()
```

## IndianFood10 Detection Dataset

The IndianFood10 dataset from Roboflow contains annotated images for object detection tasks. Each image includes bounding box annotations in YOLO format.

```python
dataset_path = roboflow_path

train_images = len(os.listdir(dataset_path + "/train/images"))
val_images = len(os.listdir(dataset_path + "/valid/images"))
test_images = len(os.listdir(dataset_path + "/test/images"))

print("Train images:", train_images)
print("Validation images:", val_images)
print("Test images:", test_images)
```

## Image Resolution Analysis

Understanding image resolution distribution helps determine the appropriate resizing strategy for model training.

```python
widths = []
heights = []

sample_images = []

for cls in khana_classes[:10]:   # sample first 10 classes
    folder = os.path.join(khana_path, cls)

    for img_file in os.listdir(folder)[:50]:
        img_path = os.path.join(folder, img_file)

        try:
            img = Image.open(img_path)
            w, h = img.size
            widths.append(w)
            heights.append(h)
        except:
            pass
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(widths)
plt.title("Image Width Distribution")

plt.subplot(1,2,2)
sns.histplot(heights)
plt.title("Image Height Distribution")

plt.show()
```

## Dataset Quality Assessment

This step checks for corrupted images and potential inconsistencies within the dataset.

```python
bad_images = []

for cls in khana_classes[:10]:
    folder = os.path.join(khana_path, cls)

    for img_file in os.listdir(folder):
        path = os.path.join(folder, img_file)

        try:
            img = Image.open(path)
            img.verify()
        except:
            bad_images.append(path)

print("Corrupted images found:", len(bad_images))
```

## Sample Image Visualization

Random images are displayed to observe dataset diversity in terms of lighting, backgrounds, and viewpoints.

```python
import random

plt.figure(figsize=(12,6))

for i in range(6):
    cls = random.choice(khana_classes)
    img_file = random.choice(os.listdir(os.path.join(khana_path, cls)))

    img = Image.open(os.path.join(khana_path, cls, img_file))

    plt.subplot(2,3,i+1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis("off")

plt.show()
```

## Image Aspect Ratio Analysis

```python
aspect_ratios = [w/h for w,h in zip(widths,heights)]

plt.figure(figsize=(6,4))
sns.histplot(aspect_ratios)

plt.title("Image Aspect Ratio Distribution")
plt.xlabel("Aspect Ratio (Width/Height)")
plt.show()
```

## Image File Size Distribution

```python
file_sizes = []

for cls in khana_classes[:10]:
    folder = os.path.join(khana_path, cls)

    for img_file in os.listdir(folder)[:50]:
        path = os.path.join(folder, img_file)
        file_sizes.append(os.path.getsize(path)/1024)

sns.histplot(file_sizes)
plt.title("Image File Size Distribution (KB)")
plt.show()
```

## Visual Inspection of Random Samples

To qualitatively understand the diversity of the dataset, a random subset of images is visualized. This step helps identify variations in food presentation, background environments, lighting conditions, and camera viewpoints. Such diversity is beneficial for training robust computer vision models.

```python
import random
import matplotlib.pyplot as plt
from PIL import Image

plt.figure(figsize=(12,8))

count = 0

while count < 9:

    cls = random.choice(khana_classes)
    img_file = random.choice(os.listdir(os.path.join(khana_path, cls)))

    img_path = os.path.join(khana_path, cls, img_file)

    try:
        img = Image.open(img_path)

        plt.subplot(3,3,count+1)
        plt.imshow(img)
        plt.title(cls)
        plt.axis("off")

        count += 1

    except:
        continue

plt.tight_layout()
plt.show()
```

## Summary of Findings

The exploratory analysis of the image datasets provided several insights into their structure and quality.

• The Khana dataset serves as the primary dataset for the project, containing over 130,000 images across 80 Indian food categories. This large dataset provides significant diversity in terms of food types and visual conditions.

• The Food20 dataset provides an additional collection of approximately 2,000 images across 20 common Indian dishes, helping increase dataset diversity and improve model generalization.

• The IndianFood10 dataset contains approximately 11,499 annotated images and will be used for training the object detection model. The dataset includes bounding box annotations in YOLO format.

• Image resolution analysis indicates that images come in different dimensions across datasets. Therefore, images will be resized to 640×640 pixels during preprocessing to ensure consistency for model training.

• The aspect ratio distribution shows moderate variation across images, indicating that food images are captured from different devices and viewpoints.

• The file size distribution demonstrates that most images fall within a moderate range, suggesting that the dataset does not contain an excessive number of extremely large files.

• Dataset size comparison highlights that the Khana dataset forms the primary training resource, while the Food20 and IndianFood10 datasets provide supplementary data for improving model robustness.

• A corrupted image check identified several unreadable images within the Khana dataset. These images will be filtered during preprocessing to ensure reliable training.

Overall, the datasets appear suitable for the NutriVision project after applying preprocessing steps such as corrupted image removal, resizing, and dataset normalization.
